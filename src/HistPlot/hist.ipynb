```python
import matplotlib.pyplot as plt
plt.rcParams.update({'text.color' : "black",
                     'axes.labelcolor' : "black"})
```

<h4><b>Plot Flle histogram <i>Before and After</i> filtering</b></h4>

```python
def hist_plot(before_path="", after_path=""):
    plt.figure(figsize=(2880*2/192, 1440*2/192), dpi=192)
    plt.style.use('fivethirtyeight')

    x_bef = []
    y_bef = []
    x_after = []
    y_after = []

    if not before_path == "":
        with open(before_path, mode='r') as f_in:
            lines = f_in.readlines()
            for line in lines:
                line = line.split()
                x_bef.append(int(line[0]))
                y_bef.append(int(line[1]))
        plt.scatter(x_bef, y_bef, c='red')

    if not after_path == "":
        with open(after_path, mode='r') as f_in:
            lines = f_in.readlines()
            for line in lines:
                line = line.split()
                x_after.append(int(line[0]))
                y_after.append(int(line[1]))
        plt.scatter(x_after, y_after, c='blue')

    plt.legend(["before", "after"], loc='upper center')
    plt.show()
```

```python
hist_plot(before_path="../../bin/before.txt", after_path="../../bin/after.txt")
```
