<h4><b><i>PLot of DCPoints that are in the same plane</i></b></h4>

```python
def equiplanar(axis1:list, axis2:list, fixed_axis:list):
    """get all points that are in the same plane

    Args:
        axis1 (list): first axis coords
        axis2 (list): second axis coords
        fixed_axis (list): fixed axis

    Returns:
        list[list[list[int]]]: each list contain points coords that are in the same plane 
    """
    fixed = []
    for _ in fixed_axis:
        if _ not in fixed:
            fixed.append(_)

    same_axis = []
    for fixed_coord in fixed:
        tmp = []
        for i in range(len(fixed_axis)):
            if fixed_axis[i] == fixed_coord:
                tmp.append([axis1[i], axis2[i]])
        same_axis.append(tmp)

    return same_axis
```

```python
plt.style.use('fivethirtyeight')
for i in range(10):
    plane = equiplanar(y, z, x)[i]

    x_coord = []; y_coord = []
    for point in plane:
        x_coord.append(point[0])
        y_coord.append(point[1])

    plt.figure(figsize=(2880*1/192, 1440*1/192), dpi=192)
    ax = plt.subplot()
    ax.scatter(x_coord, y_coord, c="#3C8BEC", s=20)
# print(y_coord)
# print(len(y_coord))
# print("------------------------------------------------")
# print(y_coord)
```
