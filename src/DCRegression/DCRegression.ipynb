<h3><b><i>Notebook to experiment Regression on DCMapped Points</i></b></h3>

```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
x = []
y = []
z = []
with open("./Points.txt", mode='r') as points:
    content = "".join(points.readlines())
    points_str = content.split()
    for point in points_str:
        point = point.replace("(", "")
        point = point.replace(")", "")
        point  = point.split(",")
        x.append(float(point[0]))
        y.append(float(point[1]))
        z.append(float(point[2]))

x_f = []
y_f = []
z_f = []
with open("./points_filtred.txt", mode='r') as points:
    content = "".join(points.readlines())
    points_str = content.split()
    for point in points_str:
        point = point.replace("(", "")
        point = point.replace(")", "")
        point  = point.split(",")
        x_f.append(float(point[0]))
        y_f.append(float(point[1]))
        z_f.append(float(point[2]))
```

```python
plt.figure(figsize=(2880*6/192, 1440*6/192), dpi=192)
plt.style.use('fivethirtyeight')

ax = plt.subplot(projection='3d')
ax.scatter3D(x, y, z, c="#3C8BEC")
ax.scatter3D(x_f, y_f, z_f, c="#FC2214")

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
# ax.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1, loc="upper left")
```
